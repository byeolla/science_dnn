```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def cos_product(N, x):
    k = np.arange(1, N+1)
    terms = 1 - 4*x**2 / (np.pi**2* (2*k -1) **2)
    return np.prod(terms)
    
```

```python
cos_product(100, 3.14)
```

```python
def factorial(N):
    prod = 1
    for i in range(2, N+1):
        prod *= i
        
    return prod
    
```

```python
factorial(5)
```

```python
def cos_taylor(N, x):
    y = np.zeros(len(x))
    for j in range(0, N+1):
        y += ((-1)**j * (x)**(2*j)/ factorial(2*j))
        # print(y)
    return y
np.array([1,2])**2
    
```

```python
x = np.linspace(-1, 3, 1000)
degree1_taylor = cos_taylor(1, x)
degree2_taylor = cos_taylor(2, x)
degree3_taylor = cos_taylor(3, x)
degree4_taylor = cos_taylor(4, x)
```

```python
plt.plot(x, np.cos(x),  = 'r', label = "cosx")
plt.plot(x, degree1_taylor, c = 'k', label = 'Degree 1')

plt.plot(x, degree2_taylor, c = 'b', label = 'Degree 2')

plt.plot(x, degree3_taylor, label = 'Degree 3')
plt.plot(x, degree4_taylor, c = 'C1', label = 'Degree 4')
plt.xlim(-1, 3)
plt.ylim(-1.5, 1.5)
plt.xlabel('$x$') ; plt.ylabel("cos")
plt.legend()
plt.show()
```
